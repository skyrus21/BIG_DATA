5)Выполнить расчет частного коэффициента корреляции между выбранными количественными показателями с исключением влияния других количественных признаков датасета.

```python
def matrix(arr, i, j):
    return np.delete(np.delete(arr,i,axis=0), j, axis=1)
def r_partial(arr, i,j):
    aij=(-1)**(i+j)*np.linalg.det(matrix(np.array(arr), i,j))
    aii=np.linalg.det(matrix(np.array(arr), i,i))
    ajj=np.linalg.det(matrix(np.array(arr), j,j))
    return -aij/math.sqrt(aii*ajj)
def corr_partial(arr):
    res=np.copy(arr)
    for i in range(np.shape(arr)[0]):
        for j in range(np.shape(arr)[0]):
            if i!=j:
                res[i,j]=r_partial(arr, i,j)
            else:
                res[i,j]=-r_partial(arr, i,j)
    res=pd.DataFrame(res)
    res.index=arr.index;res.columns=arr.columns
    return res

MyData = pd.DataFrame({'1st':a, '2nd':a1}).corr()
display(corr_partial(MyData).iloc[1, 0])
```

6)Выполнить расчет ранговых коэффициентов корреляции Спирмена и Кендалла для двух выбранных количественных показателей, оценить их значимость.

```python
Sp = data[['sum','p2']].corr('spearman')
Sp.style.background_gradient()
print("Спирмен (встроенное значение)",Sp.iloc[1, 0])
Ken = data[['sum','p2']].corr('kendall')
Ken.style.background_gradient()
print("Кендалл (встроенное значение)",Ken.iloc[1, 0])
```

Коэффициент Кендалла

```python
def MyKendall(dataFrame,columns):
    M=0
    N=0

    DR = dataFrame[columns].rank()
    DRS = DR.sort_values(by=columns[0])
    
    for i in range(DRS.shape[0]):
        df_1 = DRS.iloc[i+1:,1]
        M+= df_1[df_1 > DRS.iloc[i,1]].count()
    
    for i in range(DRS.shape[0]):
        df_1 = DRS.iloc[i+1:,1]
        N+= df_1[df_1 < DRS.iloc[i,1]].count()
    
    F=M-N
    n = DRS.shape[0]
    
    return 2*F/(n*(n-1))

print('Рассчитанный коэффициент Кендалла', MyKendall(data,['sum','p2']))
```

Коэффициент Спирмена

```python
DR = data[['sum','p2']].rank()
def Tx_Ty(series):
    FL = np.array(list(Counter(series).values()))
    T = 0.5*(FL**3-FL).sum()
    return T

ch1 = 1/6*(DR.shape[0]**3 - DR.shape[0])
ch2 = ((DR['sum']-DR['p2'])**2).sum()
Tx = Tx_Ty(data['sum'])
Ty = Tx_Ty(data['p2'])

spear_corr = (ch1 - ch2 -Tx - Ty)/(math.sqrt((ch1 - 2*Tx)*(ch1 - 2*Ty)))
T_кр = t.ppf(1 - 0.05, DR.shape[0] - 2)*math.sqrt((1 - spear_corr**2)/(DR.shape[0] - 2))
    
print('Рассчитанный коэффициент Спирмена', spear_corr)
print('Критическое значение',T_кр)
```
